```python
import xarray as xr
import rasterio 
import numpy as np
import matplotlib.pyplot as plt
```

```python
dem_cri=xr.open_rasterio('./0data/0dem/wrl_dem_cri1.tif')
```

```python
dem_cri.plot()
```

```python
dem_cri
```

```python
lon_bins = np.arange(98.0,102.0,0.1)
dem_avg = dem_cri.groupby("x").mean(dim=('y'))
```

```python
dem_avg.plot() 
```

```python
d_cut = dem_cri.sel(y=slice(20.28,20.24))
```

```python
d_cut.plot()
```

```python
d_cut_avg = d_cut.groupby("x").mean(dim=('y'))
d_cut_avg.plot() 
```

```python
d_cut_avg
```

```python
np.size(np.array(d_cut_avg))
```

```python
hgt=np.array(d_cut_avg)[0]
```

```python
dist=np.linspace(97.584583,102.174583,np.size(np.array(d_cut_avg)))
dist
```

```python
fig = plt.figure(figsize=(15,5))
ax = plt.plot(dist, hgt, color='black' )
plt.xlabel("Distance [Degree]", fontsize=25)
plt.ylabel("Height above MSL [m]", fontsize=25)
#plt.suptitle("Profile-averaged height ", fontsize=30)
#plt.tight_layout()
plt.fill_between(dist, hgt, edgecolor='silver', facecolor = 'silver')
plt.xlim(99.4994,100.839) #-40km  +100km
plt.ylim(0,1400) #-40km  +100km
plt.axvline(x=100.406, color='red',ls='--',lw='3.0' ) #
plt.axvline(x=99.881, color='black',ls='--',lw='3.0' ) #

plt.tick_params(axis='both', which='major', labelsize=25)

plt.show()
```
